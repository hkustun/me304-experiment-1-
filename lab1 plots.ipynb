```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
dfs = pd.read_excel("data1.xlsx")
```

PREPARING DATA

```python
dfs.drop(0,inplace=True)
dfs.reset_index(drop=True, inplace=True)
dfs.columns = ['Input', 'O1_up', 'O1_down', 'O2_up','O2_down','O3_up','O3_down']
dfs
```

CALCULATING AVERAGE VALUES

```python
dfs["average_up"] = (dfs['O1_up'] + dfs['O2_up'] + dfs['O3_up']) / 3
dfs["average_down"] = (dfs['O1_down'] + dfs['O2_down'] + dfs['O3_down']) / 3
dfs["total_average"] = ((dfs['O1_up'] + dfs['O2_up'] + dfs['O3_up']) + (dfs['O1_down'] + dfs['O2_down'] + dfs['O3_down'])) / 6
dfs
```

PLOTTING ALL THE DATA POINTS

```python
plt.scatter(dfs["Input"], dfs["O1_up"], color='red', label='Cycle1_up',marker='o')
plt.scatter(dfs["Input"], dfs["O2_up"], color='red', label='Cycle2_up',marker='*')
plt.scatter(dfs["Input"], dfs["O3_up"], color='red', label='Cycle3_up',marker='^')

plt.scatter(dfs["Input"], dfs["O1_down"], color='blue', label='Cycle1_down',marker='o')
plt.scatter(dfs["Input"], dfs["O2_down"], color='blue', label='Cycle2_down',marker='*')
plt.scatter(dfs["Input"], dfs["O3_down"], color='blue', label='Cycle3_down',marker='^')
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.legend()
plt.title("ALL POINTS")
plt.show()
```

AVERAGE CALCULATION PLOTS

```python
plt.scatter(dfs["Input"], dfs["average_up"], label='average_up',marker='o')
plt.scatter(dfs["Input"], dfs["average_down"], label='average_down',marker='*')
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.legend()
plt.title("AVERAGE UP-DOWN VALUES")
plt.show()
```

```python
plt.scatter(dfs["Input"], dfs["total_average"], label='average')
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.legend()
plt.title("AVERAGE VALUES")
plt.show()
```

CALCULATING LSL

```python
#create inputarray as [0.7,0,7,0.7,0,7,0.7,0.7,0.75,0.75,0.75,0.75,0.75,0.75,...] so on
#output array as [O1up,O2down,O2up,O2down,O3up,O3down,O1up,O2down,O2up,O2down,O3up,O3down,...] so on
inputarray = []
outputarray = []
for index,row in dfs.iterrows():
    for i in range(6):
        inputarray.append(row[0])
    for i in range(6):
        outputarray.append(row[i+1])
input_data2 = np.array(inputarray)
measurement_data2 = np.array(outputarray)
input_data = dfs["Input"].values
dfs["Input"] = pd.to_numeric(dfs["Input"])
dfs["total_average"] = pd.to_numeric(dfs["total_average"])
#find slope and intercept of the line
slope,intercept = np.polyfit(input_data2,measurement_data2, 1)

# Create the least squares line
line = slope * input_data + intercept

# Plot the data points and the least squares line
plt.scatter(input_data, dfs["average_up"], label='average_up')
plt.scatter(input_data, dfs["average_down"], label='average_down')
plt.plot(input_data, line, color='black', label='Least Squares Line')

# Add labels and legend
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.title('Least Squares Line with average points')
plt.legend()

# Show the plot
plt.show()
```

```python
plt.scatter(dfs["Input"], dfs["O1_up"], color='red', label='Cycle1_up',marker='o')
plt.scatter(dfs["Input"], dfs["O2_up"], color='red', label='Cycle2_up',marker='*')
plt.scatter(dfs["Input"], dfs["O3_up"], color='red', label='Cycle3_up',marker='^')

plt.scatter(dfs["Input"], dfs["O1_down"], color='blue', label='Cycle1_down',marker='o')
plt.scatter(dfs["Input"], dfs["O2_down"], color='blue', label='Cycle2_down',marker='*')
plt.scatter(dfs["Input"], dfs["O3_down"], color='blue', label='Cycle3_down',marker='^')
plt.plot(input_data, line, color='black', label='Least Squares Line')

# Add labels and legend
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.title('Least Squares Line with All Data')
plt.legend()

# Show the plot
plt.show()
```

CALCULATING MAX DEVIATIONS

```python
dfs["LSL"] = dfs["Input"] * slope + intercept
max_values = dfs[["O1_up", "O2_up", "O3_up", "O1_down", "O2_down", "O3_down"]].max(axis=1)
min_values = dfs[["O1_up", "O2_up", "O3_up", "O1_down", "O2_down", "O3_down"]].min(axis=1)
dfs["Max deviation from LSL"] = max_values - dfs["LSL"]
dfs["Min deviation from LSL"] = min_values - dfs["LSL"]
dfs["Max deviation from average"] = max_values - dfs["total_average"]
dfs["Min deviation from average"] = min_values - dfs["total_average"]
dfs
```

DEVIATION PLOTS

```python
plt.scatter(dfs["Input"], dfs["Max deviation from LSL"], label='max-deviation',marker='o')
plt.scatter(dfs["Input"], dfs["Min deviation from LSL"], label='min-deviation',marker='*')
plt.axhline(y=max(dfs["Max deviation from LSL"]), color='green', linestyle=':')
plt.axhline(y=min(dfs["Min deviation from LSL"]), color='green', linestyle=':')
plt.xlabel('Input(mm)')
plt.ylabel('Deviation(kΩ)')
plt.legend()
plt.title("DEVIATIONS FROM LSL")
plt.show()
```

```python
plt.scatter(dfs["Input"], dfs["Max deviation from average"], label='max-deviation',marker='o')
plt.scatter(dfs["Input"], dfs["Min deviation from average"], label='min-deviation',marker='*')
plt.axhline(y=max(dfs["Max deviation from average"]), color='green', linestyle=':')
plt.axhline(y=min(dfs["Min deviation from average"]), color='green', linestyle=':')
plt.xlabel('Input(mm)')
plt.ylabel('Deviation(kΩ)')
plt.legend()
plt.title("DEVIATIONS FROM AVERAGE")
plt.show()
```

ACCURACY AND PRECISION ERROR PERCENTAGES

```python
deviation_positive_accuracy = max(dfs["Max deviation from LSL"])
deviation_negative_accuracy = min(dfs["Min deviation from LSL"])
deviation_positive_precision = max(dfs["Max deviation from average"])
deviation_negative_precision = min(dfs["Min deviation from average"])
output_range = dfs.at[24,"LSL"] - dfs.at[0,"LSL"]
positive_accuracy_error = deviation_positive_accuracy / output_range * 100
negative_accuracy_error = abs(deviation_negative_accuracy) / output_range * 100
positive_precision_error = deviation_positive_precision / output_range * 100
negative_precision_error = abs(deviation_negative_precision) / output_range * 100
print("positive accuracy error percentage:","%" + str(positive_accuracy_error))
print("negative accuracy error percentage as absolute value:","%" + str(negative_accuracy_error))
print("positive precision error percentage:","%" + str(positive_precision_error))
print("negative precision error percentage as absolute value:","%" + str(negative_precision_error))
```

REPEATABILITY AND HYSTERESIS CALCULATIONS

```python
for index,row in dfs.iterrows(): #repeatability calculations 
    dfs.at[index,"repeatability"] = max((max(row[2],row[4],row[6]) - min(row[2],row[4],row[6])),(max(row[1],row[3],row[5]) - min(row[1],row[3],row[5])))
```

```python
for index,row in dfs.iterrows(): #hysteresis calculations
    dfs.at[index,"hysteresis"] = max(abs(max(row[2],row[4],row[6]) - min(row[1],row[3],row[5])),abs(max(row[1],row[3],row[5])-min(row[2],row[4],row[6])))
```

HYSTERESIS AND REPEATABILITY ERROR CALCULATIONS

```python
max_hysteresis = max(dfs["hysteresis"])
max_repeatability = max(dfs["repeatability"])
hysteresis_error = max_hysteresis/output_range * 100
repeatability_error = max_repeatability/output_range * 100
print("hysteresis error percentage:","%" + str(hysteresis_error))
print("repeatability error percentage:","%" + str(repeatability_error))
```

TERMINAL LINEARITY PLOTTING

```python
x1,x2 = dfs.at[0,"Input"], dfs.at[24,"Input"]
y1,y2 = dfs.at[0,"total_average"], dfs.at[24,"total_average"]
slope2, intercept2 = np.polyfit([x1, x2], [y1, y2], 1)
input_data = dfs["Input"].values
line2 = slope2 * input_data + intercept2
plt.plot(input_data, line2, color='black', label='terminal line')
plt.scatter(dfs["Input"], dfs["total_average"], label='average',marker='*')

# Add labels and legend
plt.xlabel('Input(mm)')
plt.ylabel('Output(kΩ)')
plt.title('Terminal Line with Average Points')
plt.legend()

# Show the plot
plt.show()
```

TERMINAL LINEARITY PERCENTAGE CALCULATION

```python
dfs["terminal line value"] = line2
dfs["terminal linearity deviation"] = abs(dfs["total_average"] - dfs["terminal line value"])
max_deviation_from_terminal_line = max(dfs["terminal linearity deviation"])
terminal_linearity_error = max_deviation_from_terminal_line/output_range * 100
print("terminal linearity percentage:", terminal_linearity_error)
```

SENSITIVITY CALCULATIONS

```python
input_range = dfs.at[24,"Input"] - dfs.at[0,"Input"] # units mm
output_range = dfs.at[24,"LSL"] - dfs.at[0,"LSL"] #units kΩ
sensitivity = output_range / input_range
print("sensitivity is:", sensitivity, "kΩ/mm")
```
